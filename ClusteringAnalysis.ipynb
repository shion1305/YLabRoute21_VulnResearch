```python
from pprint import pprint
from DataGrouper import *

clusteredData = {}
dataS = groupBySignature(data)
for s, d in dataS.items():
    sS = s.split('/')
    ws = clusteredData
    for t in sS[1:]:
        if not t in ws:
            ws[t] = {}
        ws = ws[t]
    if not '!END!' in ws:
        ws['!END!'] = 0
    ws['!END!'] += len(d)
    # ws['!!stop'].append(d)
pprint(clusteredData['recordings'])
```

```python
def repeat_print(t, level):
    for p in t.keys():
        indent = ' ' * level
        if p == '!END!':
            print(indent + "-" + str(t[p]))
            continue
        print(indent + p)
        repeat_print(t[p], level + 2)


repeat_print(clusteredData, 0)
```

半数以上はまとめることができず残っている中で、少数だが良くまとまっているものが多く存在する。

```python
def recursiveCount(target):
    rV = 0
    if type(target) == str:
        return 1
    for t in target.keys():
        if t == '!END!':
            rV += 1
            continue
        rV += recursiveCount(target[t])
    return rV


counter = {}
for d in clusteredData.keys():
    if (recursiveCount(clusteredData[d])) == 24:
        print(d)
    if not str(recursiveCount(clusteredData[d])) in counter:
        counter[str(recursiveCount(clusteredData[d]))] = []
    counter[str(recursiveCount(clusteredData[d]))].append(d)
pprint(dict(sorted(counter.items(), key=lambda item: int(item[0]))))
sorted(counter["22"])
```
